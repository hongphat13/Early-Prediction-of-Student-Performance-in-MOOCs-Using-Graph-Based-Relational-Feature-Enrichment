```python
import pandas as pd

df_phase1 = pd.read_csv("user_train_phase_1.csv")
df_phase2 = pd.read_csv("user_train_phase_2.csv")
df_phase3 = pd.read_csv("user_train_phase_3.csv")
df_phase4 = pd.read_csv("user_train_phase_4.csv")
```

## Kiểm tra cấu trúc bộ dữ liệu

```python
# Dictionary chứa các DataFrame theo phase
dfs = {
    "Phase 1": df_phase1,
    "Phase 2": df_phase2,
    "Phase 3": df_phase3,
    "Phase 4": df_phase4
}

# Lặp qua từng phase để in info và null counts
for phase_name, df in dfs.items():
    print(f"--- {phase_name} Info ---")
    df.info()
    print(f"\n--- {phase_name} Null Counts ---")
    print(df.isnull().sum())
    print("\n" + "-"*50 + "\n")
```

```python
# Categorize features
general_features = ['user_id', 'school', 'course_id', 'encoded_field_sum', 'end_year', 'user_month', 'video_count',
                    'exercise_count', 'chapter_count', 'user_past_course_count', 'user_time_since_last_course',
                    'num_prerequisites', 'certificate', 'assignment', 'video', 'exam', 'duration_days',
                    'remaining_time', 'total_score', 'label', 'label_encoded']

embedding_features = [f'combined_emb_{i}' for i in range(16)]

# DataFrames theo phase
dfs = {
    "Phase 1": df_phase1,
    "Phase 2": df_phase2,
    "Phase 3": df_phase3,
    "Phase 4": df_phase4
}

# Tạo phase-specific features
phase_specific_features = {}

for i, (phase_name, df) in enumerate(dfs.items(), start=1):
    suffix = f'_phase{i}'
    numeric_suffix = f'_{i}'
    specific_cols = [col for col in df.columns
                     if (suffix in col or numeric_suffix in col)
                     and col not in general_features + embedding_features]
    phase_specific_features[phase_name] = specific_cols

# In kết quả phân loại feature
print("--- Feature Categorization ---")
print("General Features:", general_features)
print("\nEmbedding Features:", embedding_features)
for phase, features in phase_specific_features.items():
    print(f"\n{phase} Specific Features:", features)

# Verify column count increase
col_counts = {phase: df.shape[1] for phase, df in dfs.items()}
print("\n--- Column Counts Across Phases ---")
print(col_counts)

# Kiểm tra số lượng cột có tăng dần theo phase không
col_count_values = list(col_counts.values())
column_increase_verified = all(x < y for x, y in zip(col_count_values, col_count_values[1:]))
print(f"\nDoes the number of columns increase across phases? {column_increase_verified}")

# Verify presence of embedding features
embedding_present = all(all(col in df.columns for col in embedding_features) for df in dfs.values())
print(f"\nAre embedding features present in all dataframes? {embedding_present}")

# Summarize missing values (nếu biết trước)
print("\n--- Summary of Missing Values ---")
for phase in dfs:
    print(f"{phase}: 'school' has significant missing values.")
```

### Gộp embedding và nhãn từ 4 phase vào một dataframe





```python
import numpy as np

embedding_features = [f'combined_emb_{i}' for i in range(16)]
embedding_data_phase1 = df_phase1[embedding_features + ['label']].copy()
embedding_data_phase1['phase'] = 1

embedding_data_phase2 = df_phase2[embedding_features + ['label']].copy()
embedding_data_phase2['phase'] = 2

embedding_data_phase3 = df_phase3[embedding_features + ['label']].copy()
embedding_data_phase3['phase'] = 3

embedding_data_phase4 = df_phase4[embedding_features + ['label']].copy()
embedding_data_phase4['phase'] = 4

combined_embedding_data = pd.concat([embedding_data_phase1, embedding_data_phase2, embedding_data_phase3, embedding_data_phase4], ignore_index=True)

display(combined_embedding_data.head())
display(combined_embedding_data.tail())
```

### Giảm chiều dữ liệu embedding bằng PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(combined_embedding_data[embedding_features])

combined_embedding_data['pca_component_1'] = pca_result[:, 0]
combined_embedding_data['pca_component_2'] = pca_result[:, 1]

display(combined_embedding_data.head())
```

### Scatter plot PCA và t-SNE theo nhãn phân loại

```python
from sklearn.manifold import TSNE

# Using a smaller subset for t-SNE due to computational cost
tsne = TSNE(n_components=2, random_state=42, n_jobs=-1, perplexity=30, learning_rate=200, n_iter=300)
# Select a smaller sample size for t-SNE to manage memory and computation time
sample_size = 50000
if combined_embedding_data.shape[0] > sample_size:
    tsne_data = combined_embedding_data.sample(n=sample_size, random_state=42).copy()
else:
    tsne_data = combined_embedding_data.copy()

tsne_result = tsne.fit_transform(tsne_data[embedding_features])

tsne_data['tsne_component_1'] = tsne_result[:, 0]
tsne_data['tsne_component_2'] = tsne_result[:, 1]

display(tsne_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# PCA Scatter Plot
plt.figure(figsize=(12, 6))
sns.scatterplot(x='pca_component_1', y='pca_component_2', hue='label', data=combined_embedding_data, palette='viridis', alpha=0.5)
plt.title('PCA of Combined Embedding Features Colored by Label (All Phases)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Label')
plt.grid(True)
plt.show()

# t-SNE Scatter Plot (using the sampled data)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='tsne_component_1', y='tsne_component_2', hue='label', data=tsne_data, palette='viridis', alpha=0.5)
plt.title('t-SNE of Combined Embedding Features Colored by Label (Sampled Data)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend(title='Label')
plt.grid(True)
plt.show()
```

```python
# Identify numerical phase-specific features for each phase
phase1_specific_numerical = [col for col in df_phase1.columns if ('_phase1' in col or '_1' in col) and df_phase1[col].dtype != 'object']
phase2_specific_numerical = [col for col in df_phase2.columns if ('_phase2' in col or '_2' in col) and df_phase2[col].dtype != 'object']
phase3_specific_numerical = [col for col in df_phase3.columns if ('_phase3' in col or '_3' in col) and df_phase3[col].dtype != 'object']
phase4_specific_numerical = [col for col in df_phase4.columns if ('_phase4' in col or '_4' in col) and df_phase4[col].dtype != 'object']

# Calculate descriptive statistics for these phase-specific features
desc_stats_phase1_specific = df_phase1[phase1_specific_numerical].describe()
desc_stats_phase2_specific = df_phase2[phase2_specific_numerical].describe()
desc_stats_phase3_specific = df_phase3[phase3_specific_numerical].describe()
desc_stats_phase4_specific = df_phase4[phase4_specific_numerical].describe()

# Identify key phase-specific features
key_phase_specific_features = [
    'exercise_correct_sum',
    'video_watched_percentage',
    'total_positive',
    'total_negative',
    'total_neutral',
    'entropy_time_comment'
]

# Filter descriptive statistics for key phase-specific features
def filter_key_stats(desc_stats_df, phase_num):
    phase_suffix = f'_{phase_num}' if phase_num > 1 else '_1' # Handle '_1' vs '_phaseN'
    if phase_num > 1:
         phase_suffix = f'_phase{phase_num}'

    filtered_stats = {}
    for feature in key_phase_specific_features:
        # Construct the potential column names for the current phase
        col_name_suffix = f'{feature}{phase_suffix}'
        col_name_num_suffix = f'{feature}_{phase_num}'

        if col_name_suffix in desc_stats_df.columns:
            filtered_stats[col_name_suffix] = desc_stats_df[col_name_suffix]
        elif col_name_num_suffix in desc_stats_df.columns:
             filtered_stats[col_name_num_suffix] = desc_stats_df[col_name_num_suffix]
        # Special handling for entropy_time_comment which uses _phaseN suffix consistently
        elif feature == 'entropy_time_comment' and f'entropy_time_comment_phase{phase_num}' in desc_stats_df.columns:
             filtered_stats[f'entropy_time_comment_phase{phase_num}'] = desc_stats_df[f'entropy_time_comment_phase{phase_num}']


    return pd.DataFrame(filtered_stats)


desc_stats_key_phase1 = filter_key_stats(desc_stats_phase1_specific, 1)
desc_stats_key_phase2 = filter_key_stats(desc_stats_phase2_specific, 2)
desc_stats_key_phase3 = filter_key_stats(desc_stats_phase3_specific, 3)
desc_stats_key_phase4 = filter_key_stats(desc_stats_phase4_specific, 4)


print("--- Descriptive Statistics for Key Phase-Specific Features (Phase 1) ---")
display(desc_stats_key_phase1)
print("\n--- Descriptive Statistics for Key Phase-Specific Features (Phase 2) ---")
display(desc_stats_key_phase2)
print("\n--- Descriptive Statistics for Key Phase-Specific Features (Phase 3) ---")
display(desc_stats_key_phase3)
print("\n--- Descriptive Statistics for Key Phase-Specific Features (Phase 4) ---")
display(desc_stats_key_phase4)
```
