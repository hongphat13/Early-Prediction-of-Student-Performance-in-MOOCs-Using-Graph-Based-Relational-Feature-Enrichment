```python
cols_to_remove = [
    "start_year", "start_month", "end_month", "user_year"
]

# Các mẫu cột có hậu tố _{i}, i từ 1 đến 4
pattern_cols = [
    "exercise_num_problem_mean_{}",
    "exercise_attempts_mean_mean_{}",
    "exercise_date_from_enroll_min_{}",
    "exercise_date_from_enroll_mean_{}",
    "exercise_date_from_enroll_max_{}",
    "exercise_context_mean_{}",
    "exercise_diff_sum_{}",
    "exercise_diff_min_{}",
    "exercise_diff_max_{}",
    "exercise_perc_goal_score_mean_{}",
    "exercise_perc_real_completed_std_{}",
    "exercise_perc_real_correct_mean_{}",
    "exercise_perc_real_correct_std_{}",
    "exercise_perc_real_score_sum_{}",
    "exercise_perc_real_score_std_{}",
    "video_watch_count_{}",
    "video_pause_avg_{}",
    "video_pause_std_{}",
    "video_rewatch_avg_{}",
    "video_rewatch_std_{}",
    "video_time_between_views_avg_{}",
    "entropy_time_{}"
]

# Tạo danh sách đầy đủ các cột cần xóa
for i in range(1, 5):
    cols_to_remove.extend([col.format(i) for col in pattern_cols])
```

```python
import pandas as pd
import os
```

```python
folder_path = '/kaggle/input/final-data/'
```

```python
# Duyệt đệ quy qua tất cả file CSV trong các thư mục con
for root, dirs, files in os.walk(folder_path):
    for file in files:
        # print(file)
        if file.endswith('.csv'):
            file_path = os.path.join(root, file)
            df = pd.read_csv(file_path)
            print(f"{file_path} before: {df.shape}")
            
            # Xóa cột nếu tồn tại
            df.drop(columns=[col for col in cols_to_remove if col in df.columns], inplace=True)

            # Áp dụng điều kiện: nếu label là 'D' và total_score >= 50 thì cập nhật
            mask = (df['label'] == 'D') & (df['total_score'] >= 50)
            
            # Cập nhật label và label_encoded
            df.loc[mask, 'label'] = 'C'
            df.loc[mask, 'label_encoded'] = 2

            print(f"{file_path} after: {df.shape}")
            # print(df.head(2))

            # Lưu file sau khi xử lý (giữ cấu trúc tên gốc)
            output_path = os.path.join("/kaggle/working", os.path.relpath(file_path, folder_path))
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            df.to_csv(output_path, index=False)
```
