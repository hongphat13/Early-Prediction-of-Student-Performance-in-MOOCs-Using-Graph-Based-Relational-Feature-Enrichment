## Phase 1

```python
grouped_lists_1 = grouped_lists[grouped_lists['exercise_date_from_enroll']<=14]
grouped_lists_1
```

```python
grouped_lists_1['exercise_hours'] = grouped_lists_1['exercise_hours'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)
```

```python
grouped_by_user_course = grouped_lists_1.groupby(['user_id', 'course_id']).agg({
    'exercise_id': 'count',
    'exercise_correct': ['sum', 'mean'],
    'exercise_num_problem': ['sum', 'mean'],
    'exercise_attempts_sum' : ['sum', 'mean'],	
    'exercise_attempts_mean' : 'mean', 
    # 'exercise_score': ['sum', 'mean'],
    # 'exercise_goal_score': ['sum', 'mean'],
    'exercise_date_from_enroll': ['min', 'mean', 'max'],
    'exercise_context': ['sum', 'mean'],
    'exercise_langugage_binary': 'mean',
    'exercise_diff': ['sum', 'mean', 'min', 'max'],
    'exercise_perc_goal_correct': ['sum', 'mean'],
    'exercise_perc_goal_score': ['sum', 'mean'],
    # 'exercise_real_num_problem': ['sum', 'mean'] ,
    # 'exercise_real_score':  ['sum', 'mean'],
    'exercise_perc_real_completed':  ['sum', 'mean', 'std'],
    'exercise_perc_real_correct':  ['sum', 'mean', 'std'],
    'exercise_perc_real_score':  ['sum', 'mean', 'std'],
    'exercise_hours':lambda x: sorted(
    sum([ast.literal_eval(i) if isinstance(i, str) else i for i in x], [])
)
}).reset_index()
grouped_by_user_course
```

```python
grouped_by_user_course.columns = [
    '_'.join(col).strip() if isinstance(col, tuple) and col[0] not in ['user_id', 'course_id'] else col[0]
    for col in grouped_by_user_course.columns
]
grouped_by_user_course
```

```python
grouped_by_user_course.rename(columns={'exercise_hours_<lambda>': 'exercise_hours'}, inplace=True)
```

```python
def compute_entropy_from_hour_bins(hour_list):
    if not isinstance(hour_list, list) or len(hour_list) == 0:
        return 0.0  # hoặc np.nan tùy vào yêu cầu

    # Convert to 24 hourly bins (0–23)
    hourly_bins = [h for h in hour_list]  # mỗi giờ là 2 khoảng 30 phút
    # Đếm số lần xuất hiện mỗi bin
    counts = pd.Series(hour_list).value_counts(normalize=True)

    # Tính entropy: E = -∑ p(xi) * log(p(xi))
    entropy = -sum(p * np.log(p) for p in counts)

    return round(entropy, 5)
```

```python
grouped_by_user_course['exercise_hour_entropy'] = grouped_by_user_course['exercise_hours'].apply(compute_entropy_from_hour_bins)
grouped_by_user_course
```

```python
grouped_by_user_course.drop(['exercise_hours'], axis = 1)
```

```python
grouped_by_user_course.to_csv('user_problem_1.csv', index= False)
```
