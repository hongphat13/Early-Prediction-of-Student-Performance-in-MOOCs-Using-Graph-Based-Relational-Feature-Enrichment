```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_graph(G, pos, node_options=None, edge_options=None):

    # Default node styling
    if not node_options:
        node_options = {
            'node_size': 800,
            'node_color': '#ADD8E6',  # light blue
            'edgecolors': 'black',
            'linewidths': 1.5,
            'alpha': 0.9
        }

    # Default edge styling
    if not edge_options:
        edge_options = {
            'width': 2,
            'alpha': 0.7,
            'edge_color': '#555'
        }

    # Draw graph without labels
    nx.draw_networkx_nodes(G, pos, **node_options)
    nx.draw_networkx_edges(G, pos, **edge_options)

    # No labels
    plt.axis('off')
    plt.gca().set_aspect('equal')
    # plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.cm as cm
import numpy as np
import networkx as nx

# Helper Functions

def create_color_map():
    # Define the colors for the colormap (from blue to orange)
    colors = ["#002049", "#FFA500"]
    return mcolors.LinearSegmentedColormap.from_list("blue_orange_cmap", colors)

def set_default_node_options():
    return {
        'node_color': '#4986e8',
        'node_size': 1000,
        'edgecolors': 'black',
        'linewidths': 2,
    }

def set_default_node_labels():
    return {
        'font_size': 10,
        'font_color': '#ffffff',
        'font_weight': 'bold',
    }

def set_default_edge_options():
    plt.figure(figsize=(10, 8))
    return {
        'edge_color': '#4986e8',
        'width': 2.5,
        'alpha': 0.7,
    }

def set_default_edge_labels(edge_labels):
    return {
        'edge_labels': edge_labels,
        'connectionstyle': 'arc3, rad=0.3',
        'font_size': 12,
        'font_color': 'orange',
        'bbox': {'alpha': 1, 'color': 'white'},
    }

def set_options_for_centrality(G, values, node_options):
    cent = np.fromiter(values, float)

    # Prevent division by zero
    if np.max(cent) == 0:
        cent = np.ones_like(cent)

    sizes = cent / np.max(cent) * 1200
    normalize = mcolors.Normalize(vmin=cent.min(), vmax=cent.max())
    colormap = create_color_map()

    scalar_mappable = cm.ScalarMappable(norm=normalize, cmap=colormap)
    scalar_mappable.set_array(cent)

    # Add colorbar for visual aid
    plt.colorbar(scalar_mappable, ax=plt.gca(), shrink=0.7)

    # Update node options
    node_options["node_size"] = sizes
    node_options["node_color"] = cent
    node_options["cmap"] = colormap

    return node_options
```

```python
def global_overlap(G, u, v):
    neighbors_u = set(G.neighbors(u))
    neighbors_v = set(G.neighbors(v))
    intersection = neighbors_u & neighbors_v
    union = neighbors_u | neighbors_v
    return len(intersection) / len(union) if union else 0.0
```

```python
for u, v in G.edges():
    G[u][v]['gno'] = global_overlap(G, u, v)
```

```python
df = pd.read_csv("/kaggle/input/final-data/phase1/user_train_phase_1.csv")
df
```

```python
import igraph as ig
import pandas as pd
from itertools import combinations

# Assume df is your input DataFrame with columns: user_id, school, course_id

# Step 1: Initialize Graph
print("Initializing graph...")
G = ig.Graph()

# Step 2: Create a unique list of user IDs
user_ids = df['user_id'].unique().tolist()
G.add_vertices(len(user_ids))
G.vs['user_id'] = user_ids
print(f"Added {len(user_ids)} nodes.")

# Step 3: Map user_id to igraph vertex index
user_id_to_index = {uid: idx for idx, uid in enumerate(user_ids)}
df['idx'] = df['user_id'].map(user_id_to_index)  # Pre-map to index for faster access

# Step 4: Create edges using sets to avoid duplicates
MAX_GROUP_SIZE = 800  # Adjust based on system capacity

edge_set = set()

# School-based edges
print("Processing school-based edges...")
for _, group in df.groupby('school'):
    indices = group['idx'].tolist()
    if len(indices) <= MAX_GROUP_SIZE:
        edge_set.update((min(u1, u2), max(u1, u2)) for u1, u2 in combinations(indices, 2))

# Course-based edges
print("Processing course-based edges...")
skipped_courses = []
for cid, group in df.groupby('course_id'):
    indices = group['idx'].tolist()
    if len(indices) <= MAX_GROUP_SIZE:
        edge_set.update((min(u1, u2), max(u1, u2)) for u1, u2 in combinations(indices, 2))
    else:
        skipped_courses.append((cid, len(indices)))

print(f"Skipped {len(skipped_courses)} course groups due to size.")


# Step 5: Add all unique edges to the graph
G.add_edges(list(edge_set))
print(f"Added {len(edge_set)} unique edges.")

# Step 6: Run Walktrap community detection
print("Running Walktrap community detection...")
walktrap = G.community_walktrap(steps=4)
clusters = walktrap.as_clustering()
print(f"Detected {len(clusters)} communities.")

# Step 7: Map vertex index to cluster
print("Mapping users to clusters...")
vertex_to_cluster = {G.vs[vid]["user_id"]: cluster_id
                     for cluster_id, community in enumerate(clusters)
                     for vid in community}

# Step 8: Assign clusters back to the original DataFrame
print("Assigning clusters to DataFrame...")
df_with_cluster = df[['user_id', 'course_id']].copy()
df_with_cluster['cluster'] = df_with_cluster['user_id'].map(vertex_to_cluster)

# Step 9: Final result
result_df = df_with_cluster[['user_id', 'course_id', 'cluster']]
print("Done.")
```

```python
# Step X: Compute closeness centrality
print("Computing closeness centrality...")
closeness_values = G.closeness()

# Map closeness centrality to user_id
user_id_to_closeness = {
    G.vs[i]['user_id']: closeness_values[i]
    for i in range(len(G.vs))
}

# Add closeness to DataFrame
df_with_cluster['closeness'] = df_with_cluster['user_id'].map(user_id_to_closeness)
result_df = df_with_cluster
```
